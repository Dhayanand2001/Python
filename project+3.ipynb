```python
#8.
import pandas as pd
data = pd.read_csv('youtube_dislike_dataset.csv')
data['published_at']=pd.to_datetime(data['published_at'])
data['published_month']=data['published_at'].dt.month
data['published_month']=data['published_month'].map(month)
COUNTS=data['published_month'].value_counts()
SCOUNTS=COUNTS.sort_values(ascending=False)
SCOUNTS
```

```python
#11.data.groupby('channel_title')['video_id'].count().sort_values(ascending=False).head(10)
pd.pivot_table(data, index = ['title'], values = ['likes'], aggfunc = np.max).sort_values(by = 'likes', ascending = False).head(10)
```

```python
#12.Find the title of the video which has the maximum number of dislikes and the title of the video having minimum dislikes and write your inferences.
pd.pivot_table(data,index=['title'],values=['dislikes'],aggfunc=np.max).sort_values(by ='dislikes', ascending = False).head(10)
```

Inference: 'Cuties | Offical Trailer | Netflix' has the highest number of dislikes.
